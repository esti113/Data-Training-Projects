```python
# key statistics for executive summary
print(""" EXECUTIVE SUMMARY """)

# Total crimes analysed
total_crimes = len(all_crime_data)
print(f"Total Crimes Analysed: {total_crimes:,}")

# Time period covered
years = all_crime_data['Year'].unique()
print(f"Time Period: {min(years)}-{max(years)}")

# Police forces covered
police_forces = all_crime_data['Falls within'].nunique()
print(f"Police Forces Analysed: {police_forces}")

# Top areas by crime count (excluding 'Unknown')
filtered_df = all_crime_data[all_crime_data['LSOA name'] != 'Unknown']
top_areas = filtered_df['LSOA name'].value_counts().head(5)
print("\nTop 5 Areas by Crime Volume (excluding 'Unknown'):")
print(top_areas)
```

```python

#regional comparison
# Comparative analysis between regions
plt.figure(figsize=(14, 7))
regional_analysis = all_crime_data['Falls within'].value_counts()

# Bar chart with annotations
ax = sns.barplot(x=regional_analysis.values, y=regional_analysis.index, palette="Blues_d")
plt.title('Total Crimes by Police Force Area', fontsize=16, pad=20)
plt.xlabel('Number of Crimes', fontsize=12)
plt.ylabel('Police Force Area', fontsize=12)

# Add value annotations
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 100, p.get_y() + p.get_height()/2., 
            f'{int(width):,}', 
            ha='left', va='center', fontsize=10)

plt.tight_layout()
plt.show()

# Crime rate per LSOA within each police force
plt.figure(figsize=(14, 8))
top_boroughs_per_force = (all_crime_data.groupby(['Falls within', 'LSOA name'])
                          .size()
                          .groupby(level=0, group_keys=False)
                          .nlargest(5)
                          .unstack()
                          .fillna(0))

top_boroughs_per_force.plot(kind='barh', stacked=True, colormap='viridis')
plt.title('Top 5 LSOAs by Crime Count in Each Police Force', fontsize=16)
plt.xlabel('Number of Crimes', fontsize=12)
plt.ylabel('Police Force Area', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Borough')
plt.tight_layout()
plt.show()
```

```python
# Crime type distribution with real estate relevance
plt.figure(figsize=(14, 8))
crime_type_dist = all_crime_data['Crime type'].value_counts()

# Create a custom color palette where violent crimes are highlighted
colors = ['#ff6b6b' if 'Violence' in x or 'Robbery' in x else '#4ecdc4' for x in crime_type_dist.index]

ax = sns.barplot(x=crime_type_dist.values, y=crime_type_dist.index, palette=colors)
plt.title('Crime Type Distribution (Violent Crimes Highlighted)', fontsize=16, pad=20)
plt.xlabel('Number of Incidents', fontsize=12)
plt.ylabel('Crime Type', fontsize=12)

# Add percentage annotations
total = float(len(all_crime_data))
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_width()/total)
    x = p.get_x() + p.get_width() + total * 0.005  # Smaller shift
    y = p.get_y() + p.get_height()/2
    ax.annotate(percentage, (x, y), va='center', fontsize=10)

plt.tight_layout()
plt.show()

# Violent crime proportion by region
violent_crimes = ['Violence and sexual offences', 'Robbery']
violent_df = all_crime_data[all_crime_data['Crime type'].isin(violent_crimes)]

plt.figure(figsize=(14, 7))
violent_by_region = (violent_df['Falls within'].value_counts() / 
                    all_crime_data['Falls within'].value_counts()).sort_values(ascending=False)

sns.barplot(x=violent_by_region.values, y=violent_by_region.index, palette="Reds_d")
plt.title('Proportion of Violent Crimes by Police Force Area', fontsize=16)
plt.xlabel('Proportion of Violent Crimes', fontsize=12)
plt.ylabel('Police Force Area', fontsize=12)
plt.xlim(0, 0.5)  # Set consistent scale for comparison
plt.tight_layout()
plt.show()
```

```python
# Monthly and yearly trends
plt.figure(figsize=(14, 8))
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

monthly_trends = all_crime_data.groupby(['Year', 'Month Name']).size().unstack().T
monthly_trends = monthly_trends.loc[month_order]

sns.lineplot(data=monthly_trends, dashes=False, markers=True, palette="husl", linewidth=2.5)
plt.title('Monthly Crime Trends by Year', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.legend(title='Year', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Yearly trends by crime type (focus on property crimes)

# define relevant property crime types
property_crimes = ['Burglary', 'Vehicle crime', 'Theft from the person', 'Other theft']

# filter  main dataframe to include only property crimes
property_df = all_crime_data[all_crime_data['Crime type'].isin(property_crimes)]

# Group by Year and Crime type
yearly_property = (
    property_df.groupby(['Year', 'Crime type'])
    .size()
    .unstack()
    .sort_index()
)

# Make sure 'Year' is treated as integer for clean x-axis
yearly_property.index = yearly_property.index.astype(int)

# Plot the line chart
plt.figure(figsize=(14, 8))
sns.lineplot(data=yearly_property, dashes=False, markers=True, linewidth=2.5)

plt.title('Yearly Trends of Theft-Related Crimes', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Crimes', fontsize=12)
plt.xticks(yearly_property.index)  # Ensure clean integer ticks on x-axis
plt.legend(title='Crime Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```

```python
# Top crime locations analysis (most/least desirable areas)
top_locations = all_crime_data['Location'].value_counts().head(20)
bottom_locations = all_crime_data['Location'].value_counts().tail(20)  # Least crime-prone

# Plot top locations
plt.figure(figsize=(14, 8))
sns.barplot(x=top_locations.values, y=top_locations.index, palette="rocket_r")
plt.title('Top 20 Highest Crime Locations', fontsize=16)
plt.xlabel('Number of Crimes', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.tight_layout()
plt.show()

# Plot bottom locations
plt.figure(figsize=(14, 8))
sns.barplot(x=bottom_locations.values, y=bottom_locations.index, palette="mako_r")
plt.title('20 Lowest Crime Locations', fontsize=16)
plt.xlabel('Number of Crimes', fontsize=12)
plt.ylabel('Location', fontsize=12)
plt.tight_layout()
plt.show()

```
