```python
 #INVADE AND CONQUER - SPACE RAIDERS STYLE GAME 


import pygame
import sys
import os
import random

# Initialise pygame
pygame.init()


# SOUND HANDLING SYSTEM - prevents the gamne from crashing even if the files are missing
class SafeSound:
    
    def __init__(self, filename=None):
        self.sound = None
        self.available = False
        try:
            if filename and os.path.exists(filename):
                self.sound = pygame.mixer.Sound(filename)
                self.available = True
        except:
            pass
    
    def play(self):
        #Play the sound if available, silently fail if not
        if self.available and self.sound:
            try:
                self.sound.play()
            except:
                pass

# Initialise game sounds with safe fallback
laser_sound = SafeSound("laser.wav")
bass_sound = SafeSound("bass.wav")
level_sound = SafeSound("levelup.wav")
explode_sound = SafeSound("explode.ogg")
game_over_sound = SafeSound("gameover.wav")


# GAME SETTINGS AND CONSTANTS

# Screen dimensions
WIDTH, HEIGHT = 900, 800
WIN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Invade and Conquer")

# Colour definitions
WHITE = (255, 255, 255)
BLUE = (0, 0, 255)
RED = (255, 0, 0)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
BLACK = (0, 0, 0)

# Font settings
FONT = pygame.font.SysFont("Arial", 40)
SMALL_FONT = pygame.font.SysFont("Arial", 24)

# Gameplay constants
PLAYER_WIDTH, PLAYER_HEIGHT = 60, 60
PLAYER_VEL = 5
BULLET_VEL = 7
ENEMY_VEL_X_START = 1
ENEMY_DROP_START = 20
ENEMY_BULLET_VEL = 4
LEVEL_UP_SPEED = 0.3
LEVEL_UP_DROP = 2
ENEMY_ROWS, ENEMY_COLS = 3, 7
ENEMY_SPACING_X, ENEMY_SPACING_Y = 100, 60
LEADERBOARD_FILE = "leaderboard.txt"

# Game clock
clock = pygame.time.Clock()
FPS = 60

# UTILITY FUNCTIONS AND CLASSES

class Button:
    """Button class for menu system"""
    def __init__(self, x, y, width, height, text):
        self.rect = pygame.Rect(x, y, width, height)
        self.text = text

    def draw(self, win):
        """Draw the button on the specified surface"""
        pygame.draw.rect(win, GREEN, self.rect)
        txt = SMALL_FONT.render(self.text, True, BLACK)
        win.blit(txt, (self.rect.centerx - txt.get_width() // 2, 
                      self.rect.centery - txt.get_height() // 2))

    def clicked(self, pos):
        """Check if button was clicked"""
        return self.rect.collidepoint(pos)

def draw_text_middle(text, font, color, surface, y_offset=0):
    """Helper to draw centered text"""
    txt = font.render(text, True, color)
    surface.blit(txt, (WIDTH // 2 - txt.get_width() // 2, 
                      HEIGHT // 2 - txt.get_height() // 2 + y_offset))

def safe_quit():
    """Safely quit the game in any environment"""
    pygame.quit()
    try:
        sys.exit()
    except:
        pass

def save_score(name, score):
    """Save player score to leaderboard file"""
    with open(LEADERBOARD_FILE, 'a') as f:
        f.write(f"{name},{score}\n")

def show_leaderboard():
    """Load and return top 5 scores from leaderboard"""
    if not os.path.exists(LEADERBOARD_FILE):
        return []
    scores = []
    with open(LEADERBOARD_FILE, 'r') as f:
        for line in f:
            parts = line.strip().split(',')
            if len(parts) == 2 and parts[1].isdigit():
                scores.append((parts[0], int(parts[1])))
    scores.sort(key=lambda x: x[1], reverse=True)
    return scores[:5]

def create_enemies():
    """Create a new grid of enemies"""
    return [pygame.Rect(100 + j * ENEMY_SPACING_X, 50 + i * ENEMY_SPACING_Y, 40, 40)
            for i in range(ENEMY_ROWS) for j in range(ENEMY_COLS)]


# SCREEN FUNCTIONS

def show_main_menu():
    """Display the main menu screen"""
    play_button = Button(WIDTH//2 - 100, HEIGHT//2 - 60, 200, 50, "Play Game")
    instructions_button = Button(WIDTH//2 - 100, HEIGHT//2, 200, 50, "Instructions")
    quit_button = Button(WIDTH//2 - 100, HEIGHT//2 + 60, 200, 50, "Quit Game")

    while True:
        WIN.fill(BLACK)
        draw_text_middle("Invade and Conquer", FONT, WHITE, WIN, -150)
        play_button.draw(WIN)
        instructions_button.draw(WIN)
        quit_button.draw(WIN)
        pygame.display.update()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                safe_quit()
            if event.type == pygame.MOUSEBUTTONDOWN:
                pos = pygame.mouse.get_pos()
                if play_button.clicked(pos):
                    player_name = get_player_name()
                    run_game(player_name)
                elif instructions_button.clicked(pos):
                    show_instructions()
                elif quit_button.clicked(pos):
                    safe_quit()

def show_instructions():
    """Display game instructions screen"""
    back_button = Button(WIDTH//2 - 100, HEIGHT - 100, 200, 50, "Go Back")

    while True:
        WIN.fill(BLACK)
        draw_text_middle("Instructions", FONT, WHITE, WIN, -300)

        instructions = [
            "Some Aliens from Planet Zorg want to invade your planet",
            "Move with left and right arrows to dodge enemy bullets",
            "Shoot with spacebar",
            "Destroy all enemies to level up",
            "Good Luck!",
        ]

        for i, line in enumerate(instructions):
            line_text = SMALL_FONT.render(line, True, WHITE)
            WIN.blit(line_text, (WIDTH//2 - line_text.get_width()//2, 200 + i * 40))

        back_button.draw(WIN)
        pygame.display.update()

        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                safe_quit()
            if event.type == pygame.MOUSEBUTTONDOWN:
                if back_button.clicked(pygame.mouse.get_pos()):
                    return

def get_player_name():
    """Screen for player to enter their name"""
    input_box = pygame.Rect(WIDTH//2 - 150, HEIGHT//2 - 20, 400, 60)
    text = ''
    warning = ""
    cursor_visible = True
    cursor_timer = 0
    flash_timer = 0

    while True:
        dt = clock.tick(60)
        cursor_timer += dt
        flash_timer += dt
        
        # Toggle cursor visibility every 500ms
        if cursor_timer >= 500:
            cursor_visible = not cursor_visible
            cursor_timer = 0
            
        WIN.fill(BLACK)
        draw_text_middle("Enter Your Name", FONT, WHITE, WIN, -100)

        # Draw text input
        txt_surface = FONT.render(text, True, WHITE)
        WIN.blit(txt_surface, (input_box.x + 5, input_box.y + 5))

        # Draw blinking cursor
        if cursor_visible:
            cursor = pygame.Rect(input_box.x + 5 + txt_surface.get_width(), 
                               input_box.y + 5, 2, txt_surface.get_height())
            pygame.draw.rect(WIN, WHITE, cursor)

        pygame.draw.rect(WIN, WHITE, input_box, 2)
        draw_text_middle("Press Enter to Start", FONT, WHITE, WIN, 100)

        # Show warning if needed
        if warning:
            warn_text = SMALL_FONT.render(warning, True, RED)
            WIN.blit(warn_text, (WIDTH//2 - warn_text.get_width() // 2, HEIGHT//2 + 100))

        pygame.display.flip()

        # Handle input events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                safe_quit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_RETURN:
                    if text.strip():
                        return text.strip()
                    else:
                        warning = "Please enter your name"
                elif event.key == pygame.K_BACKSPACE:
                    text = text[:-1]
                else:
                    text += event.unicode

def draw_window(player, enemies, bullets, enemy_bullets, score, lives, level):
    """Draw all game elements to the screen"""
    WIN.fill(BLACK)
    
    # Draw player
    pygame.draw.rect(WIN, BLUE, player)
    
    # Draw bullets
    for bullet in bullets:
        pygame.draw.rect(WIN, YELLOW, bullet)
    
    # Draw enemies
    for enemy in enemies:
        pygame.draw.circle(WIN, GREEN, (enemy.x + 20, enemy.y + 20), 20)
    
    # Draw enemy bullets
    for bullet in enemy_bullets:
        pygame.draw.rect(WIN, RED, bullet)

    # Draw HUD
    WIN.blit(SMALL_FONT.render(f"Score: {score}", True, WHITE), (10, 10))
    WIN.blit(SMALL_FONT.render(f"Lives: {lives}", True, WHITE), (10, 40))
    WIN.blit(SMALL_FONT.render(f"Level: {level}", True, WHITE), (10, 70))

    pygame.display.update()

def display_game_over(player_name, score):
    """Show game over screen with leaderboard"""
    replay_button = Button(WIDTH // 2 - 100, HEIGHT // 2 + 100, 200, 50, "Play Again")
    quit_button = Button(WIDTH // 2 - 100, HEIGHT // 2 + 160, 200, 50, "Quit Game")
    
    while True:
        WIN.fill(BLACK)
        
        # Game over text
        draw_text_middle("GAME OVER", FONT, RED, WIN, -250)
        draw_text_middle(f"Score: {score}", SMALL_FONT, WHITE, WIN, -200)

        # Show leaderboard
        leaderboard = show_leaderboard()
        y = HEIGHT//2 - 150
        WIN.blit(SMALL_FONT.render("Top 5 Players:", True, GREEN), (WIDTH//2 - 100, y))
        
        for i, (name, scr) in enumerate(leaderboard):
            y += 30
            WIN.blit(SMALL_FONT.render(f"{i+1}. {name}: {scr}", True, WHITE), 
                    (WIDTH//2 - 100, y))

        # Draw buttons
        replay_button.draw(WIN)
        quit_button.draw(WIN)
        pygame.display.update()

        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                safe_quit()
                return False
                
            if event.type == pygame.MOUSEBUTTONDOWN:
                pos = pygame.mouse.get_pos()
                if replay_button.clicked(pos):
                    return True  # Play again
                elif quit_button.clicked(pos):
                    safe_quit()
                    return False  # Quit game


# MAIN GAME LOGIC

def run_game(player_name):
    """Main game loop"""
    # Initialise game state
    score = 0
    lives = 3
    level = 1
    ENEMY_VEL_X = ENEMY_VEL_X_START
    ENEMY_DROP = ENEMY_DROP_START
    
    # Create game objects
    player = pygame.Rect(WIDTH // 2 - PLAYER_WIDTH // 2, 
                        HEIGHT - PLAYER_HEIGHT - 10, 
                        PLAYER_WIDTH, PLAYER_HEIGHT)
    bullets = []
    enemy_bullets = []
    enemies = create_enemies()
    enemy_dir = 1
    can_shoot = True

    # Main game loop
    while lives > 0 and level <= 10:
        clock.tick(FPS)
        
        # Handle events
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                safe_quit()
                return
                
        # Player movement
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT] and player.x - PLAYER_VEL > 0:
            player.x -= PLAYER_VEL
        if keys[pygame.K_RIGHT] and player.x + PLAYER_WIDTH + PLAYER_VEL < WIDTH:
            player.x += PLAYER_VEL
            
        # Player shooting
        if keys[pygame.K_SPACE] and can_shoot and len(bullets) < 5:
            bullets.append(pygame.Rect(player.x + PLAYER_WIDTH // 2 - 5, player.y, 10, 20))
            laser_sound.play()
            can_shoot = False
            
        if not keys[pygame.K_SPACE]:
            can_shoot = True

        # Move player bullets and check for hits
        for bullet in bullets[:]:
            bullet.y -= BULLET_VEL
            if bullet.y < 0:
                bullets.remove(bullet)
                continue
                
            for enemy in enemies[:]:
                if bullet.colliderect(enemy):
                    enemies.remove(enemy)
                    if bullet in bullets:
                        bullets.remove(bullet)
                    score += 10
                    explode_sound.play()
                    break

        # Move enemy bullets and check for player hit
        for bullet in enemy_bullets[:]:
            bullet.y += ENEMY_BULLET_VEL
            if bullet.y > HEIGHT:
                enemy_bullets.remove(bullet)
            elif bullet.colliderect(player):
                enemy_bullets.remove(bullet)
                lives -= 1
                bass_sound.play()
                if lives <= 0:
                    game_over_sound.play()
                    if display_game_over(player_name, score):
                        run_game(player_name)  # Restart game
                    return

        # Move enemies
        for enemy in enemies:
            enemy.x += ENEMY_VEL_X * enemy_dir
            
            # Check if enemies hit screen edge
            if enemy.x <= 0 or enemy.x + 40 >= WIDTH:
                enemy_dir *= -1
                for e in enemies:
                    e.y += ENEMY_DROP
                break

        # Check if enemies reached bottom
        for enemy in enemies:
            if enemy.y + 40 >= HEIGHT:
                game_over_sound.play()
                if display_game_over(player_name, score):
                    run_game(player_name)  # Restart game
                return

        # Enemy random shooting
        if random.random() < 0.02 and enemies:
            shooting_enemy = random.choice(enemies)
            enemy_bullets.append(pygame.Rect(shooting_enemy.x + 20, shooting_enemy.y + 40, 5, 15))

        # Level up if all enemies destroyed
        if not enemies:
            level += 1
            level_sound.play()
            if level <= 10:
                ENEMY_VEL_X += LEVEL_UP_SPEED
                if ENEMY_DROP < 30:
                    ENEMY_DROP += LEVEL_UP_DROP
                
                draw_text_middle(f"Level {level}", FONT, WHITE, WIN)
                pygame.display.update()
                pygame.time.delay(1500)
                
                # Reset for next level
                bullets.clear()
                enemy_bullets.clear()
                enemies = create_enemies()

        # Draw everything
        draw_window(player, enemies, bullets, enemy_bullets, score, lives, level)

    # Save score when game ends
    save_score(player_name, score)
    if display_game_over(player_name, score):
        run_game(player_name)  # Restart if player chooses to play again


# START THE GAME

if __name__ == "__main__":
    show_main_menu()

```
